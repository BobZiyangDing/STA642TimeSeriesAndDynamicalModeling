* (a)

We know that:
\begin{aligned}
    y_{t-1} &\sim \mathrm{N}(x_{t-1}, w)\\
    x_t &\sim \mathrm{N}(x_{t-1} \phi, v)\\
    y_t &\sim \mathrm{N}(x_{t-1} \phi, v + w)\\
    \phi y_{t-1} &\sim \mathrm{N}(x_{t-1}\phi, \phi^2 w)\\
    \phi y_{t-1} + \epsilon + \nu_t - \phi \nu_{t-1} &\sim \mathrm{N}(x_{t-1}\phi , \phi^2 w + v + w - \phi^2 w)\\
    &\sim \mathrm{N}(x_{t-1} \phi, v + w)\\
    &= y_{t} = \phi y_{t-1} + \eta_t
\end{aligned}

* (b)

\begin{aligned}
    \mathrm{Cor}(\eta_{t}, \eta_{t-1}) &=\mathrm{Cor}(\epsilon_{t}+\nu_{t}-\phi \nu_{t-1}, \epsilon_{t-1}+\nu_{t-1}-\phi \nu_{t-2})\\
    &= \frac{\mathrm{Cov}(\epsilon_{t}+\nu_{t}-\phi \nu_{t-1}, \epsilon_{t-1}+\nu_{t-1}-\phi \nu_{t-2})}{\sqrt{\mathrm{Var}(\epsilon_{t}+\nu_{t}-\phi \nu_{t-1})\mathrm{Var}(\epsilon_{t}+\nu_{t}-\phi \nu_{t-1})}}\\
    &= \frac{\mathrm{E}[(\epsilon_{t}+\nu_{t}-\phi \nu_{t-1})(\epsilon_{t-1}+\nu_{t-1}-\phi \nu_{t-2})]  }{v+w + \phi^2 w}\\
    &= \frac{\mathrm{E}[\epsilon_t\epsilon_{t-1} + \epsilon_t \nu_{t-1} - \phi \epsilon_t \nu_{t-2} ] }{v+w + \phi^2 w}\\
    &\quad + \frac{\mathrm{E}[\nu_t\epsilon_{t-1} + \nu_t \nu_{t-1} - \phi \nu_t \nu_{t-2}]  }{v+w + \phi^2 w}\\
    &\quad - \frac{\mathrm{E}[\phi\nu_{t-1}\epsilon_{t-1} + \phi\nu_{t-1} \nu_{t-1} - \phi^2\nu_{t-1} \nu_{t-2} ] }{v+w + \phi^2 w}\\
    &= -\frac{\phi w}{v+w + \phi^2 w}\\
    &= \frac{-\phi w}{v+(1+\phi^2)w}\\
\end{aligned}
